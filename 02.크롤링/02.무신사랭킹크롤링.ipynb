▶️ 전체 베스트 상품 필요정보 데이터프레임화

```python
item_code = []
item_url_list = []
item_brand_list = []
item_name_list = []
item_price_list = []
item_review_cnt_list = []


for best_outer in best_list:
    product_url = best_outer.select_one("p.list_info > a").get("href")
    parsed_url = urlparse(product_url)
    url_code2 = parsed_url.path
    goodscode = url_code2.split("goods/")[1]
    brand_name = best_outer.select_one("div.article_info > p.item_title").text.replace("\n","")
    product_name = best_outer.select_one("p.list_info").text.replace("\n","").replace(" ","")
    price = best_outer.select_one("div.article_info > p.price").text.split()[-1].replace(",","").replace("원","")
    review_cnt = best_outer.select_one("div.article_info > p.point > span.count")
    product_url = best_outer.select_one("p.list_info > a").get("href")

    if review_cnt == None :
        review_cnt = 0
    
    else:
        review_cnt = best_outer.select_one("div.article_info > p.point > span.count").text
    
    
    item_code.append(goodscode)
    item_url_list.append(product_url)
    item_brand_list.append(brand_name)
    item_name_list.append(product_name)
    item_price_list.append(price)
    item_review_cnt_list.append(review_cnt)
```

```python
datas = {
    "상품코드" : item_code,
    "브랜드": item_brand_list,
    "제품명": item_name_list,
    "상품가격": item_price_list,
    "상품리뷰수": item_review_cnt_list,
    "링크": item_url_list
}

df = pd.DataFrame(datas)
df.head(5)
```

```python
%%writefile musinsa.py
import requests
from bs4 import BeautifulSoup

all_page_products = []

def get_total_page():
    url = "https://search.musinsa.com/ranking/best?u_cat_cd=002"
    html_page = requests.get(url).content
    soup = BeautifulSoup(html_page, 'html.parser')
    total_page = soup.select_one("#goodsRankForm > div.right_contents.hover_box > div.boxed-list-wrapper.rank-shop > div.thumbType_box.box > span.pagingNumber > span.totalPagingNum").text
    total_page = int(total_page)
    return total_page

def get_musinsa_product(category="002"):
    total_page = get_total_page()
    for page in range(1, total_page+1):
        url = "https://search.musinsa.com/ranking/best?mainCategory=002&page={}".format(page)

        html_page = requests.get(url.format(page)).content
        soup = BeautifulSoup(html_page, 'html.parser')
        products = soup.select("#goodsRankList > li")
        print("Page {} Crwaling...".format(page))
        for product in products:

            # 상품 id
            product_id = product['data-goods-no']
            img_url = product.select_one("img.lazyload").get("data-original")
            product_brand = product.select_one("p.item_title").text
            product_detail_url = product.select_one("p.list_info > a").get("href")
            product_name = product.select_one("p.list_info").text.strip()
            product_price = product.select_one("p.price").text.split()
            product_price = float(product_price[-1].replace(",", "").replace("원", ""))
            start_count_elem = product.select_one("p.point > span.count")
            if start_count_elem:
                star_count = int(start_count_elem.text.replace(",", ""))
            else:
                star_count = 0

            item = {
                "product_id": product_id,
                "img_url": img_url,
                "product_id":product_id,
                "img_url":img_url,
                "product_brand":product_brand,
                "product_detail_url":product_detail_url,
                "product_name":product_name,
                "product_price":product_price,
                "star_count":star_count
            }

            all_page_products.append(item)
            
    return all_page_products
```
