## 파이썬에서 가장 기본적인 시각화 도구
데이터를 숫자 형식으로 보는 것보다 시각적으로 추세나 분포 등을 보는 것이 더 효율적

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

1. 밑그림 그리기 

```python
data = np.arange(1, 100)
```

```python
plt.figure() # 캔버스가 생성, 처음 시각화할때는 안 써줘도 된다.

plt.plot(data) # data를 토대로 시각화 진행

plt.show() # 그래프 보여주기
```

```python
# 캔버스 없이 시각화를 하면 자동으로 캔버스가 만들어진다.
plt.plot(data)
plt.show()
```

```python
# 캔버스 사이즈 조절하기
plt.figure(figsize=(12,6)) # figsize = (가로크기, 세로크기) 단위는 인치
plt.plot(data)
plt.show()
```

캔버스 하나에 그래프 하나가 그려지면 **단일그래프** 💚💜💜💜 중요

```python
# 캔버스 하나에 여러 개의 그래프 그려보기
plt.figure() # 맨 앞에 썼으면 굳이 안써도 됨
data1 = np.arange(1,51)
data2 = np.arange(51,101)

plt.plot(data1)
plt.plot(data2)

plt.show()
```

```python
plt.figure() # 맨 앞에 썼으면 굳이 안써도 됨
plt.plot(data1)

plt.figure()
plt.plot(data2)

plt.show()
```

# 하나의 캔버스에 여러 개의  plot을 그리기
* subplot(row, column, index)
    * 인덱스는 1부터 시작
    * 인덱스로 지정한 위치에 그래프를 그림
   

```python
plt.subplot(2, 1, 1) # row 2개 컬럼 1개 인덱스는 하나 2행 1열에 첫번째
plt.plot(data1)

plt.subplot(2, 1, 2) #plt.subplot(212) 이렇게 써도 가능
                    # 2행 1열에 두번째
plt.plot(data2)

plt.show()
```

```python
plt.subplot(1, 2, 1) # 1행 2열에 첫번째
plt.plot(data1)

plt.subplot(1, 2, 2)# 1행 2열에 두번째 (2번째 열)
plt.plot(data2)

plt.show()
```

# 여러 개의 plot을 그리는 방법 2
* subplots

```python
fig, axes = plt.subplots(2, 3)
fig.set_figheight(8) # 가로 세로 높이 조절 
# or 위에서 plt.subplots(2, 3, figsize=(20,10)) 이렇게 해도 됨
fig.set_figwidth(15)

axes[0,0].plot(data1)
axes[0,1].plot(data1 * data1)
axes[0,2].plot(data1**3)
axes[1,0].plot(data % 10)
axes[1,1].plot(-1 * data1)
axes[1,2].plot(data // 20)
plt.show()
```
