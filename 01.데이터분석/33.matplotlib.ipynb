```python
# Ï∫îÎ≤ÑÏä§ ÌïòÎÇòÏóê Ïó¨Îü¨ Í∞úÏùò Í∑∏ÎûòÌîÑ Í∑∏Î†§Î≥¥Í∏∞
plt.figure() # Îß® ÏïûÏóê ÏçºÏúºÎ©¥ Íµ≥Ïù¥ ÏïàÏç®ÎèÑ Îê®
data1 = np.arange(1,51)
data2 = np.arange(51,101)

plt.plot(data1)
plt.plot(data2)

plt.show()
```

# ÌïòÎÇòÏùò Ï∫îÎ≤ÑÏä§Ïóê Ïó¨Îü¨ Í∞úÏùò  plotÏùÑ Í∑∏Î¶¨Í∏∞
* subplot(row, column, index)
    * Ïù∏Îç±Ïä§Îäî 1Î∂ÄÌÑ∞ ÏãúÏûë
    * Ïù∏Îç±Ïä§Î°ú ÏßÄÏ†ïÌïú ÏúÑÏπòÏóê Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
   

```python
plt.subplot(2, 1, 1) # row 2Í∞ú Ïª¨Îüº 1Í∞ú Ïù∏Îç±Ïä§Îäî ÌïòÎÇò 2Ìñâ 1Ïó¥Ïóê Ï≤´Î≤àÏß∏
plt.plot(data1)

plt.subplot(2, 1, 2) #plt.subplot(212) Ïù¥Î†áÍ≤å Ïç®ÎèÑ Í∞ÄÎä•
                    # 2Ìñâ 1Ïó¥Ïóê ÎëêÎ≤àÏß∏
plt.plot(data2)

plt.show()
```

```python
plt.subplot(1, 2, 1) # 1Ìñâ 2Ïó¥Ïóê Ï≤´Î≤àÏß∏
plt.plot(data1)

plt.subplot(1, 2, 2)# 1Ìñâ 2Ïó¥Ïóê ÎëêÎ≤àÏß∏ (2Î≤àÏß∏ Ïó¥)
plt.plot(data2)

plt.show()
```

# Ïó¨Îü¨ Í∞úÏùò plotÏùÑ Í∑∏Î¶¨Îäî Î∞©Î≤ï 2
* subplots

```python
fig, axes = plt.subplots(2, 3)
fig.set_figheight(8) # Í∞ÄÎ°ú ÏÑ∏Î°ú ÎÜíÏù¥ Ï°∞Ï†à 
# or ÏúÑÏóêÏÑú plt.subplots(2, 3, figsize=(20,10)) Ïù¥Î†áÍ≤å Ìï¥ÎèÑ Îê®
fig.set_figwidth(15)

axes[0,0].plot(data1)
axes[0,1].plot(data1 * data1)
axes[0,2].plot(data1**3)
axes[1,0].plot(data % 10)
axes[1,1].plot(-1 * data1)
axes[1,2].plot(data // 20)
plt.show()
```
