```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('2021_1분기보고서_01_재무상태표_20220215.txt', sep='\t', encoding='euc-kr')
df1.head(3)
```

```python
df = df1[['종목코드','업종','업종명','항목명', '당기 1분기말']]
df.head(5)
```

```python
df.isnull().sum()
```

```python
# dropna() - 결측치가 포함된 행을 삭제
df.dropna(inplace=True)
df.shape
```

```python
# 자산 삭제
df.isnull().sum()
```

```python
df.head()
```

```python
df.info()
```

```python
df['당기 1분기말'] = df['당기 1분기말'].apply(lambda x: int(x.replace(',','')))
df.head(3)
```

```python
df2 = df.copy()
df2 = df2.drop(df2[df2.업종 != 261].index)
df2.head()
```

```python
df2['종목코드'] = df2['종목코드'].apply(lambda x: str(x.replace('[','')))
df2['종목코드'] = df2['종목코드'].apply(lambda x: str(x.replace(']','')))
df2.head(100)
```
