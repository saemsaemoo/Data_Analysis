```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('2020_1분기보고서_01_재무상태표_연결_20211104.txt', sep='\t', encoding='euc-kr')
df1.head(3)
```

```python
df = df1[['종목코드','업종','업종명','항목명', '당기 1분기말']]
df.head(5)
```

```python
df.isnull().sum()
```

```python
# dropna() - 결측치가 포함된 행을 삭제
df.dropna(inplace=True)
df.shape
```

```python
# 자산 삭제
df.isnull().sum()
```

```python
df.head()
```

```python
df.info()
```

```python
df['당기 1분기말'] = df['당기 1분기말'].apply(lambda x: int(x.replace(',','')))
df['종목코드'] = df['종목코드'].apply(lambda x: str(x.replace('[','')))
df['종목코드'] = df['종목코드'].apply(lambda x: str(x.replace(']','')))
df['항목명'] = df['항목명'].apply(lambda x: str(x.strip()))

df.head(3)
```

```python
df2 = df.copy()
df2 = df2.drop(df2[df2.업종 != 261].index)
df2.head()
```

```python
df2.info()
```

```python
asset_mask = df2['항목명'].str.contains('유동자산')
money_mask = df2['항목명'].str.contains('현금및현금성자산')
acc_receivable_mask = df2['항목명'].str.contains('매출채권')
ivt_mask = df2['항목명'] .str.contains('재고자산')
nonivt_mask = df2['항목명'] .str.contains('비유동자산')
capital_mask = df2['항목명'] .str.contains('자본금')
df_flow2 = df2.loc[asset_mask|money_mask|acc_receivable_mask|ivt_mask|nonivt_mask|capital_mask]
df_flow2.head(20)
```
