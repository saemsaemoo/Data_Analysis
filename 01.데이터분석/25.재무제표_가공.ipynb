```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./data/재무제표파일/반도체기업_2021_4분기사업보고서.csv')
```

```python
df['corp_code'].nunique()
```

```python
df_con= df[df['fs_nm'] == '연결재무제표']
df_sep = df[df['fs_nm'] == '재무제표']
```

```python
list = df_con['corp_code'].unique().tolist()
df_sep = df_sep[df_sep['corp_code'].isin(list) == False]
df_sep['corp_code'].nunique()
df = pd.concat([df_con, df_sep])
df
```

```python
df = df[['corp_code', 'fs_nm', 'account_nm', 'thstrm_amount', 'thstrm_dt']]
df = df[df['account_nm'].isin(['유동자산', '자산총계', '유동부채', '부채총계', '자본금', '이익잉여금', '자본총계', '매출액', '영업이익', '당기순이익'])]
df = df.reset_index(drop=True)
df
```

```python
# 이렇게 할 수도 있다
df.pivot(index='corp_code', columns = 'account_nm', values = 'thstrm_amount')
```

```python
df1 = pd.pivot_table(df, index='corp_code', columns = 'account_nm', values = 'thstrm_amount', aggfunc=np.sum, fill_value=0)
df1 = df1.reset_index()
df1
```

```python
df_aaa = df1[['corp_code', '당기순이익', '매출액', '영업이익']]
df_aaa.head()
```
