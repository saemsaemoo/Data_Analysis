```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df2017 = pd.read_csv('./data/재무제표파일/반도체_2017가공1.csv')
df2018 = pd.read_csv('./data/재무제표파일/반도체_2018가공1.csv')
df2019 = pd.read_csv('./data/재무제표파일/반도체_2019가공1.csv')
df2020 = pd.read_csv('./data/재무제표파일/반도체_2020가공1.csv')
df2021 = pd.read_csv('./data/재무제표파일/반도체_2021가공1.csv')
```

```python
df2017sales = df2017[['corp_code', '매출액']]
df2017sales = df2017sales.rename(columns={'매출액':'2017'})
df2017sales.head()
```

```python
df2018sales = df2018[['corp_code', '매출액']]
df2018sales = df2018sales.rename(columns={'매출액':'2018'})
df2018sales.head()
```

```python
df2019sales = df2019[['corp_code', '매출액']]
df2019sales = df2019sales.rename(columns={'매출액':'2019'})
df2019sales.head()
```

```python
df2020sales = df2020[['corp_code', '매출액']]
df2020sales = df2020sales.rename(columns={'매출액':'2020'})
df2020sales.head()
```

```python
df2021sales = df2021[['corp_code', '매출액']]
df2021sales = df2021sales.rename(columns={'매출액':'2021'})
df2021sales.head()
```

```python
df_merge1 = pd.merge(df2017sales,df2018sales, on='corp_code', how='right')
df_merge1.head()
```

```python
df_merge2 = pd.merge(df2019sales,df2020sales, on='corp_code', how='right')
df_merge2.head()
```

```python
df_merge3 = pd.merge(df_merge1, df_merge2, on='corp_code', how='right')
df_merge3.head()
```

```python
df_net = pd.merge(df_merge3, df2021sales, on='corp_code', how='right')
df_net.head()
```

```python
df_profit = df_net.set_index('corp_code')
df_profit.head()
```
