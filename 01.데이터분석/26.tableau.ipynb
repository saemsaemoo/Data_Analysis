```python
df2017dang = df2017[['corp_code', '당기순이익']]
df2017dang = df2017dang.rename(columns={'당기순이익':'2017'})
df2017dang
```

```python
df2018dang = df2018[['corp_code', '당기순이익']]
df2018dang = df2018dang.rename(columns={'당기순이익':'2018'})
df2018dang
```

```python
df2019dang = df2019[['corp_code', '당기순이익']]
df2019dang = df2019dang.rename(columns={'당기순이익':'2019'})
df2019dang
```

```python
df2020dang = df2020[['corp_code', '당기순이익']]
df2020dang = df2020dang.rename(columns={'당기순이익':'2020'})
df2020dang
```

```python
df2021dang = df2021[['corp_code', '당기순이익']]
df2021dang = df2021dang.rename(columns={'당기순이익':'2021'})
df2021dang
```

```python
df_merge1 = pd.merge(df2017dang,df2018dang, on='corp_code', how='right')
df_merge1.head()
```

```python
df_merge2 = pd.merge(df2019dang,df2020dang, on='corp_code', how='right')
df_merge2.head()
```

```python
df_merge3 = pd.merge(df_merge1, df_merge2, on='corp_code', how='right')
df_merge3.head()
```

```python
df_merge3.info()
```

```python
df_net = pd.merge(df_merge3, df2021dang, on='corp_code', how='right')
df_net.head()
```

```python
df_net = df_net.set_index('corp_code')
df_net.head()
```
